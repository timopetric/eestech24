```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from constants import *
from glob import glob
from datetime import datetime
import os

df_prazniki = pd.read_csv(DATA_PRAZNIKI, sep=';', encoding='UTF-8')
df_prazniki.columns = df_prazniki.columns.str.lower()
df_prazniki['datum'] = pd.to_datetime(df_prazniki['datum'], format='%d.%m.%Y')

# keep only datum, dela_prost_dan
df_prazniki = df_prazniki[['datum', 'dela_prost_dan']]

# take only 2022, 2023, 2024
df_prazniki = df_prazniki[df_prazniki['datum'].dt.year >= 2022]
df_prazniki = df_prazniki[df_prazniki['datum'].dt.year <= 2024]

# strip and map da ne to True False
df_prazniki['dela_prost_dan'] = df_prazniki['dela_prost_dan'].str.strip()
assert df_prazniki['dela_prost_dan'].unique().shape[0] == 2, f"unique = {df_prazniki['dela_prost_dan'].unique()}"
df_prazniki['dela_prost_dan'] = df_prazniki['dela_prost_dan'].map({'da': True, 'ne': False})

# sort by datum
df_prazniki = df_prazniki.sort_values(by=['datum'])

# save to csv
df_prazniki.to_csv(OUT_PRAZNIKI, index=False)

df_prazniki = pd.read_csv(OUT_PRAZNIKI)
df_prazniki
```
